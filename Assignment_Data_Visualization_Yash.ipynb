```python
#Que 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap?

#Ans. # Generate synthetic data with correlated features
np.random.seed(0)
n_samples = 100
x1 = np.random.normal(0, 1, n_samples)
x2 = 0.8 * x1 + np.random.normal(0, 0.2, n_samples)
x3 = -0.5 * x1 + 0.6 * x2 + np.random.normal(0, 0.3, n_samples)
data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#Que 4. Using the given x and y data, generate a 3D surface plot to visualize the function  using plotly
# z= sin(underootx^2+y^2)
# x = np.linspace(-5, 5, 100)
# y = np.linspace(-5, 5, 100)
# x, y = np.meshgrid(x, y)
# z = np.sin(np.sqrt(x**2 + y**2))
# data = {
#     'X': x.flatten(),
#     'Y': y.flatten(),
#     'Z': z.flatten()
# }
# df = pd.DataFrame(data)

import plotly.graph_objects as go


# Given data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Customize the plot
fig.update_layout(title='3D Surface Plot', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

# Display the plot
fig.show()
```

```python
#BOKEH ASSIGNMENT:

#Que 1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
import numpy as np
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

# Create x values
x = np.linspace(0, 10, 100)

# Calculate y values (sine of x)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Add a line glyph
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

```python
#Que 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the  markers based on the 'sizes' and 'colors' columns

import numpy as np
from bokeh.plotting import figure, show

# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(10, 50, 50)
colors = ["red" if i < 25 else "blue" for i in range(50)]

# Create a Bokeh figure
p = figure(title="Scatter Plot with Varying Sizes and Colors",
           x_axis_label="X", y_axis_label="Y")

# Add scatter glyphs with varying sizes and colors
p.circle(x, y, size=sizes, color=colors, alpha=0.8)

# Show the plot
show(p)


```

```python
#Que 3.Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add bar glyphs
p.vbar(x=fruits, top=counts, width=0.5)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
#Que 4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

from bokeh.models import ColumnDataSource

# Create a Bokeh figure
p = figure(title="Histogram", x_axis_label="Values", y_axis_label="Frequency")

# Create a ColumnDataSource from the histogram data
source = ColumnDataSource(data=dict(hist=hist, left=edges[:-1], right=edges[1:]))

# Add a quad glyph to represent the histogram
p.quad(bottom=0, top='hist', left='left', right='right', source=source, fill_color="navy", line_color="white")

# Show the plot
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)


from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, LinearColorMapper
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()


# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=xx.flatten(), y=yy.flatten(), z=data_heatmap.flatten()))

# Create a Bokeh figure
p = figure(title="Bokeh Heatmap", x_axis_label="X", y_axis_label="Y")

# Add a heatmap glyph
# Create a LinearColorMapper instance
mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

# Use the mapper in the fill_color dictionary
p.rect(x="x", y="y", width=0.1, height=0.1, source=source,
       fill_color={'field': 'z', 'transform': mapper},
       line_color=None)

# Show the plot
show(p)
```
